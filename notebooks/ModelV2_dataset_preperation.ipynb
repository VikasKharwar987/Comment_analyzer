# Processing train.csv

```python
df3 = pd.read_csv("train.csv")
df3.shape
```

```python
df3.head()
```

```python
toxicity_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
df3['label'] = df3[toxicity_cols].any(axis=1).astype(int)

print(df3['label'].value_counts())
display(df3.head())
```

```python
df3 = df3[["comment_text", "label"]]
df3["text"] = df3["comment_text"]
df3 = df3[["text", "label"]]
```

```python
df3.head()
```

```python
import pandas as pd
from sklearn.utils import resample
jigsaw_safe = df3[df3["label"] == 0]
jigsaw_toxic = df3[df3["label"] == 1]

target = len(jigsaw_toxic)

jigsaw_safe_bal = resample(
    jigsaw_safe,
    replace=False,
    n_samples=target,
    random_state=42
)

jigsaw_bal = pd.concat([jigsaw_safe_bal, jigsaw_toxic])
jigsaw_bal = jigsaw_bal.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
print(jigsaw_bal["label"].value_counts())
```

```python
combined = pd.concat([df, df1, df2, jigsaw_bal], ignore_index=True)
```

```python
combined.to_csv("final_dataset.csv", index=False)
```
